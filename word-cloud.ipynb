```python
from os import path
import pickle
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
t160 = 'tmp/160_topics'
t220 = 'tmp/220_topics'
t340 = 'tmp/340_topics'
t60 = 'tmp/60_topics'
t120 = 'tmp/120_topics'
topic_files = [t120]
wc = WordCloud(background_color="#EDEDED", width=1200, height=800, font_path='/Users/Muk/Library/Fonts/Helvetica.dfont')
```

```python
for tf in topic_files:
    with open(tf, 'rb') as f:
        topics = pickle.load(f)
        for t, topic in enumerate(topics):
            words = [(j,i) for i, j in topic]
            wc.fit_words(words)
            plt.imshow(wc)
            #plt.axis("off")
            #plt.show()
            wc.to_file( 'visualization/' + str(tf.split("/")[1]) + '/t' + str(t) + '.png' )
```
